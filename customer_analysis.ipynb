# Data quality and cleaning

**Displaying the count and percentage of null values**

```python
# Calculate the number and percentage of null values for each column
null_counts = df.isnull().sum()
null_percentage = (null_counts / len(df)) * 100


# Create a DataFrame to summarize the null count and null percentage
null_summary = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentage})
print(null_summary[null_summary['Null Count'] > 0])
```

**Fill the missing values with mean**

```python
numerical_columns = df.select_dtypes(include=np.number).columns

for column in numerical_columns:
    # exception handling
    try:
       # Calculate the average and fill in missing values
        mean_value = df[column].mean()
        df[column] = df[column].fillna(mean_value)
    except TypeError as e:
        print(f"TypeError for column '{column}': {e}")
    except Exception as e:
        print(f"An error occurred with column '{column}': {e}")

# Check missing value statistics
null_summary_after = pd.DataFrame({
    'Null Count': df.isnull().sum(),
    'Null Percentage': (df.isnull().sum() / len(df)) * 100
})
print(null_summary_after[null_summary_after['Null Count'] > 0])

```

**Identifing duplicate columns and rows**

```python
# Calculate the number of duplicated rows before removing columns
duplicate_count = df.duplicated().sum()
print(f"The dataset contains {duplicate_count} duplicated rows.")
```

# Data transformation

**Transform Data Type**

```python
# Perform date format conversion
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format="%d-%m-%Y")
df['Complain'] = df['Complain'].astype('category')
df['Response'] = df['Response'].astype('category')


print("Changing Dt_Customer's type from object to", df['Dt_Customer'].dtype)
print("Changing Complain's type from int64  to", df['Complain'].dtype)
print("Changing Response's type from int64  to", df['Response'].dtype)

```

**Rename Column**

```python
# Rename the column to Customer_enrolldate
df.rename(columns={'Dt_Customer': 'Customer_enrolldate'}, inplace=True)

print("changing name of Dt_Customer to Customer_enrolldate")
```

**Creating some columns**

```python
# Calculate the total number of children in the household
df['totalchildrenhome'] = df['Kidhome'] + df['Teenhome']

# Calculate total expenses across different product categories
df['Expenses'] = df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']

# Calculate total accepted campaign across different campaign
df['TotalAcceptedCmp'] = df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5']

# Calculate the total number of purchases made by the customer
df['TotalNumPurchases'] = df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']

# Calculate the average expenses by the customer
df['AverageExpenses']=df['Expenses'] / df['TotalNumPurchases']

# calculate the customers' age at 2021 year
dataset_year = 2021
df['Age'] = dataset_year - df['Year_Birth']

# Print the first few rows of the newly created columns
new_columns = ['totalchildrenhome', 'Expenses', 'TotalAcceptedCmp','TotalNumPurchases','AverageExpenses','Age']
print(df[new_columns].head())
```

**Remove unnecessary columns**

```python
# Remove unnecessary columns
df.drop(['Year_Birth','ID', 'Z_CostContact', 'Z_Revenue',"AcceptedCmp1" , "AcceptedCmp2", "AcceptedCmp3" , "AcceptedCmp4","AcceptedCmp5",
         "NumWebPurchases","NumCatalogPurchases","NumStorePurchases" , "Kidhome", "Teenhome"], axis=1, inplace=True)
print("Remaining these variables: ", df.columns)
```

**Data replacement**

```python
replace_marital_status = {
    'Single': 'Single',
    'Married': 'Married',
    'Divorced': 'Divorced',
    'Widow': 'Widowed',
    'Together': 'In a Relationship',
    'Alone': 'Single',
    'Absurd': 'Other',
    'YOLO': 'Other'
}

df['Marital_Status'] = df['Marital_Status'].replace(replace_marital_status)
print("Marital_Status after replacing: ", df['Marital_Status'].unique())

```

**Converting all values of string columns to lowercase**

```python
# Filter string columns
string_columns = df.select_dtypes(include='object').columns

for col in string_columns:
    # Convert to lowercase if the column contains any alphabetic characters
    if df[col].str.contains('[a-zA-Z]').any():
        df[col] = df[col].str.lower()
        print(f"Column '{col}' has been converted to lowercase:\n{df[col].unique()}\n")
    else:
        print(f"Column '{col}' is excluded from lowercase conversion.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**check Outliers**

```python
columns_to_check = ['Income','Expenses','AverageExpenses','Age']

# Loop through numerical columns and plot vertical box plots to detect outliers
for column in columns_to_check:
    plt.figure(figsize=(5, 5))
    sns.boxplot(y=df[column])
    plt.title(f"Box Plot for {column}")
    plt.show()

```

```python
# Function to remove outliers from specified columns
print("Original shape:", df.shape)

def remove_outliers(df, columns):
    cleaned_df = df.copy()
    for col in columns:
        Q1 = np.percentile(cleaned_df[col], 25)
        Q3 = np.percentile(cleaned_df[col], 75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        cleaned_df = cleaned_df[(cleaned_df[col] >= lower_bound) & (cleaned_df[col] <= upper_bound)]
    return cleaned_df

# Remove outliers from specified columns
df_cleaned = remove_outliers(df, columns_to_check)
print("Shape after removing outliers:", df_cleaned.shape)
```
